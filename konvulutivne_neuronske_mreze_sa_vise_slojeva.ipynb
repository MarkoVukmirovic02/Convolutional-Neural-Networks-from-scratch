Aktivacione funkcije i njihove derivacije

```python
def sigmoid(Z):
  A=1/(1+np.exp(-Z))
  return A

def softmax(Z):
  expZ=np.exp(Z)
  return expZ/(np.sum(expZ,0))

def relu(Z):
  A= np.maximum(0,Z)
  return A

def tanh(x):
  return np.tanh(x)

def derivative_relu(Z):
  return np.array(Z>0,dtype='float')

def derivative_tanh(x):
  return (1-np.power(x,2))
```

Maxpooling metoda backward pass

```python
def max_pooling_backward(dA_pool, pool_cache):
# ideja je da se vratimo na pocetne dimenzije matrice koja je bila pre nego što smo vršili metodu max pool forward
    pool_size=pool_cache['pool_size'] #iz informacija pool_cache koju smo dobili i sacuvali kad smo radili max pool forward
    stride=pool_cache['stride']# vadimo informacije o tome koliki je bio pool  koliki korak
    switches=pool_cache['switches']#informacije o lokacijama maximalnih vrednost
    H,W,C=pool_cache['A_shape'] # dimenzije pocetne matrice tj matrice koja ce ovde biti izlaz a bila je ulaz u forward pasu

    dA=np.zeros((H,W,C),dtype=dA_pool.dtype) # kreiramo nule tih dimenzija tj dimenzija izlazne matrice

    Hp,Wp,C_check=dA_pool.shape # dimenzije poolovane matrice  tj matrice koja je bila prodakt poolinga a ovde je ulazna informacija

    assert C_check == C, f"Channel mismatch: dA_pool has C={C_check}, expected {C}" #


# krecemo se po dimenzijama poolovane matrice
    for oi in range(Hp):
      for oj in range(Wp):
        for c in range(C):
          i_star=switches[oi,oj,c,0] # vadimo informacije o lokacijama prisetimo se switches cuva vrednosti (i*,j*) gde cuva lokaciju maksimalnog elementa tog poola
          j_star=switches[oi,oj,c,1]
          dA[i_star,j_star,c]+=dA_pool[oi,oj,c] # u novu matricu zappisujemo tu vrednost koja se nalazi na oi,oj lokaciji ali se originalno nalazi na (i*,j*)

    return dA # vracamo ovu matricu koja je bila originalna pre poolinga imamo sada i njene dimenzije i njene najvece najbitnije vrednosti po pooling prozoru
```

Maxpooling metoda forward pass

```python
def max_pool_forward(A, pool_size=2, stride=2):
    # A: (H, W, C)
    H, W, C = A.shape  # uzimamo dimenzije da bi mogli izracunati dimenzije izlazne matrice
    Hp = (H - pool_size) // stride + 1  # koristimo poznatu formulu H=floor((Hprev-f)/stride) +1
    Wp = (W - pool_size) // stride + 1
    out = np.zeros((Hp, Wp, C), dtype=A.dtype)   # pravimo dummy matricu koju cemo popunjavati

    # switches nam cuvaju informaciju o lokaciji tj indeksima maksimalnog elementa
    switches = np.zeros((Hp, Wp, C, 2), dtype=int)  # (oi, oj, c) -> (i*, j*)

    oi = 0
    for i in range(0, H - pool_size + 1, stride): # maksimalno se pomeramo od 0 do dimenzije matrice A - te dimenzije poola  korakom stride se pomeramo
        oj = 0
        for j in range(0, W - pool_size + 1, stride):
            window = A[i:i+pool_size, j:j+pool_size, :]      # uzmemo odredjen prozor flattenujemo ga nadjemo maximum i index maximuma
            flat = window.reshape(pool_size*pool_size, C)    # za odgovarajucu dimenziju C uzmemo prozor
            idx = np.argmax(flat, axis=0)                    # uzmemo max
            di = idx // pool_size                            # nadjemo gde se nalazi u samom prozoru
            dj = idx %  pool_size
            out[oi, oj, :] = flat[idx, np.arange(C)]     # zapisemo vrednost u out matricu tj maximalnu vrednost
            switches[oi, oj, :, 0] = i + di              # zapisujemo u switches matricu stvarne dimenzije tj lokacije maximuma u originalnoj matrici
            switches[oi, oj, :, 1] = j + dj
            oj += 1
        oi += 1

    pool_cache = {'pool_size': pool_size, 'stride': stride,
                  'switches': switches, 'A_shape': A.shape}
    return out, pool_cache
```

Forward Propagacija klasicne neuronske mreze

```python
def forward_propagation(X,parametars,activation):

  forward_cache={}# recnik u kome cuvamo vrednosti numeracija ce kretati od A0, Z1,A1.....ZL, AL
  L=len(parametars)//2  # parametri su recnik {w1: ,b1:,..wL:,bL} celobrojnim deljenjem sa 2 dobijamo broj slojeva
  forward_cache['A0']=X#  U suštini ovo je prvi ulazni vektor tj ovo je vektor koji dobijamo flattenovanjem ALc tj nakon primene aktivacione
  #funkcije na rezultat poslednjeg konvulucionog sloja

  for l in range(1,L):# za prvih 1,....L-1 racunamo po formuli koju smo izveli
    forward_cache['Z'+str(l)]=parametars['W'+str(l)].dot(forward_cache['A'+str(l-1)])+parametars['b'+str(l)]
# U zavisnosti od odgovarajuce aktivacione funkcije
    if activation=='tanh':
      forward_cache['A'+str(l)]=tanh(forward_cache['Z'+str(l)])
    else:
      forward_cache['A'+str(l)]=relu(forward_cache['Z'+str(l)])
# Poslednji ZL koji nam je bitan jer nakon primene jedne od klasifikacionih aktivacionih fja on postaje finalna predikcija
  forward_cache['Z'+str(L)]=parametars['W'+str(L)].dot(forward_cache['A'+str(L-1)])+parametars['b'+str(L)]
# u zavisnosti od prirode problema binarna sigmoid, multiklasna softmax
  if forward_cache['Z'+str(L)].shape[0]==1:
    forward_cache['A'+str(L)]=sigmoid(forward_cache['Z'+str(L)])
  else:
    forward_cache['A'+str(L)]=softmax(forward_cache['Z'+str(L)])
  return forward_cache['A'+str(L)],forward_cache # vracamo finalnu predikciju dimenzija (1,1), (10,1) i čitav forward_cache
  # dimenzije su (1,1) (10,1)  nas algoritam radi samo na pojedinacnim slikama posle cemo ih grupisati funkcijom forward_propagation_batch
```

```python
def forward_propagation_conv(# parametri koje prima funkcija su X- pocetna slika dimenzija (1,64,64,3) kao što smo rekli u suštini mi radimo sa pojedinačnim
    x, conv_layers, conv_params, fc_params,#slikama pa ih vezujemo putem forward_propagation_batch funkcije, conv_params, fc_params za svaki naredni poziv  forward porpagacije
    activation='relu', stride=1, pooling=True, pool_every=2# jer je bitno da napredujemo, aktivaciona funkcija koja se koristi, stride nam govori koliko se filter pomera elemenata u desno i dole
):# prilikom vršenja operacije konvulucije, pooling =True da li da radimo pooling i pool_every na koliko iteracija da ga radimo
    cache = {} # recnik u kome skladistimo A0,Z1,A1,...ZLc,ALc
    cache['A0'] = x #kao sto smo rekli pocetni vektor
    A_prev = x
    Lc = len(conv_layers) #broj slojeva
# conv_lazers je lista torki koja sadrzi sledece informacije  K-broj filtera, f_h-prvu dimenziju filtera, f_w-drugu dimenziju filtera, C_in-trecu dimenziju filtera
    for l in range(Lc):
        K, f_h, f_w, C_in = conv_layers[l]
        H_in, W_in, C_prev = A_prev.shape  #Ove vrednosti su nam neophodne da bi izracunali vrednost dimenzija Z_next matrice tj nju cemo prvo staviti dummy da su sve nule
        assert C_prev == C_in, f"Channel mismatch: got {C_prev}, expected {C_in} for layer {l+1}"  # pa cemo je popunjavati redom

        H_out = (H_in - f_h) // stride + 1  # racun dimenzija po formulama koje imamo
        W_out = (W_in - f_w) // stride + 1
        Z = np.zeros((H_out, W_out, K), dtype=A_prev.dtype)  # pravljenje dummy matrice

        # Primena konvulucije najpre prolazimo kroz sve filtere ! bitno je napomenuti da nam indeksiranje nije u skladu
        for k in range(K): # naime parametri prvog filtera koji je filter 0 tj to je njegov indeks
            Wk = conv_params[l]['w' + str(k+1)]   # ima odgovarajuce parametre w1 ,b1 to je zato što smo stavili indeksiranje K od 0 da krece do K-1
            bk = conv_params[l]['b' + str(k+1)]   # Dok indeksiranje W krece od 1 do K
            oi = 0   # sada je ideja primene kliznog prozora prisetimo se naš filter prolazi kroz naš A_prev skenira jedan deo uradi konvuluciju i sabere sve vrednosti u tom prozoru
            for i in range(0, H_in - f_h + 1, stride):# i krece od 0 maksimalno moze da se pomeri do pozicije duzina matrice A - duzina filtera, stride nam govori korak sa kojim se pomeramo
                oj = 0   # poziciranje oi i oj su mesta na Z 0_dummy matrici na koja upisujemo resenje
                for j in range(0, W_in - f_w + 1, stride): # analog gornjem objasnjenju
                    patch = A_prev[i:i+f_h, j:j+f_w, :]  # uzmemo odredjeni isecak koji je dimenzija  kao i sam filter dakle [f_hxf_w]
                    Z[oi, oj, k] = np.sum(patch * Wk) + bk  # mnozimo sa Wk i sabiramo sve pa na to dodajemo bias kao resenje dobijamo vrednost Z pozicija oi,oj u k toj dimenziji
                    oj += 1   # setimo se primenom k filtera nas izlaz ce imati 3. dimenziju velicine k
                oi += 1

        # u zavisnosti od aktivacione funkcije
        if activation == 'tanh':
            A = np.tanh(Z)
        else:
            A = np.maximum(0, Z) # tanh ili Relu se primenjuje

        # Metoda pooling , kao sto smo objasnili cela ideja je smanjivanje kompleksnosti samog modela na taj nacin da se osigurava da informacija ostaje prenesena
        did_pool = False # bilo da radimo ili ne  postavljamo False
        pool_cache = None # vrednosti koje skladistimo u pool_cache su veličina poola, stride ili korak, switches matrica koja sadrzi informacije o poziciji maksimalnog elementa (i*,j*,c*) i izlazne dimenzije AL
        if pooling and ((l + 1) % pool_every == 0): # ako je pooling True i ako je red da se primenjuje u trenutnom sloju
            # primenjujemo max pool funkciju za forward propagaciju
            A, pool_cache = max_pool_forward(A, pool_size=2, stride=2)
            did_pool = True  # odradili smo pooling

        # Zapisujemo u naš keš
        cache['Z' + str(l+1)] = Z
        cache['A' + str(l+1)] = A
        cache['meta' + str(l+1)] = {
            'f_h': f_h, 'f_w': f_w, 'stride': stride,
            'did_pool': did_pool, 'pool': pool_cache
        }

        A_prev = A  # postavljamo sadasnji za prethodni i ponavljamo petlju
# nakon zavrsene petlje izlazni AL je ustvari produkt poslednje konvulucije i ustvari njega treba da flattenujemo u vektor i prosledimo fully connected layeru
    conv_out = A_prev
    cache['conv_out_shape'] = conv_out.shape # pamtimo njegove dimenzije da bi mogli da se vratimo nazad u Backpropagaciji

    # Ako ne koristimo prvi put funkciju onda radimo forward propagaciju da dobijemo izlazni tenzor i forwarc cache
    if fc_params is not None:
        A0_flat = conv_out.reshape(-1, 1)
        AL, fc_cache = forward_propagation(A0_flat, fc_params, activation=activation)
    else:
        AL, fc_cache = conv_out, None

    return AL, fc_cache, cache  # u slucaju da ne primenjujemo prvi put dobijamo izlazni tenzor predikcija
```

```python
import numpy as np

def initialize_parameters(conv_layers, input_shape, fc_layers,
                          pooling=True, pool_every=2,
                          activation='relu', stride=1, seed=None):
    """
    Funkcija koja inicijalizuje parametre konvulucijskih slojeva i fully connected slojeva.

    Argumenti:
        conv_layers: lista torki (K, f_h, f_w, C_in) za svaki  pojedinacni conv layer.
        input_shape: (H, W, C) dimenzije ulazne slike.
        fc_layers: lista intigera za FC layers (svaki broj oznacava broj neurona u datom sloju
        poslednji broj je broj neurona poslednjeg sloja i on mora da se poklapa sa brojem kategorija koje predvidjamo).
        pooling: da li primenjujemo pooling na konvolucionim slojevima.
        pool_every: ako primenjujemo, primenjujemo svaki N-ti konvulucioni sloj.
        activation: aktivaciona funkcija konvulucionog sloja (Relu Tanh).
        stride: conv stride tj koliko se prozor odnosno filter pomera po matrici Al (ovde se predpostavlja da je isti za sve conv layere).
        seed: koristi se za slucajan izbor parametra

    Returns:
        conv_params: recnik parametara po konvulucijonom sloju,: {'w1','b1',...,'wK','bK'}
        fc_params: recnik of parametara po FC sloju : {'W1','b1',...}
    """
    if seed is not None:
        np.random.seed(seed)

    conv_params = {}

    # ---- Convolutional params ----
    # Kao sto smo rekli conv_layers je torka K- broj filtera  f_h,f_w,C_in dimenzije filtera
    for l, (K, f_h, f_w, C_in) in enumerate(conv_layers):
        layer_params = {} # parametri u tom layeru
        # Tipičan izbor standardizacija
        #He init for ReLU, small normal for tanh:
        scale = 0.01 if activation == 'tanh' else np.sqrt(2.0 / (f_h * f_w * C_in))
        for k in range(K): # prolazimo po Filterima u datom sloju i za odgovarajuci filter belezimo njegove vrednosti tj parametre.
            layer_params[f'w{k+1}'] = np.random.randn(f_h, f_w, C_in) * scale  #prisetimo se opet zbog numeracije 0-ti filter ima parametre w1,b1
            layer_params[f'b{k+1}'] = 0.0  # k ide od 0 do K-1   a vrednosti parametara su od 1 do K
        conv_params[l] = layer_params # dodajemo recnik u recnik

    # ---- Dakle ideja je da napravimo dummy tenzor dimenzija pocetnog inputa tj slike koju ubacujemo (no FC) ----
    dummy_input = np.zeros(input_shape, dtype=float)  # zatim da je provucemo kroz forward propagaciju
    AL, fc_cache_dummy, conv_cache = forward_propagation_conv(# na taj nacin dobijamo dimenzije tenzora nakon primene poslednje konvulucije i aktivacione funkcije
        x=dummy_input,# tj dobijamo  dimenzije tenzora koji se flattenuje
        conv_layers=conv_layers,
        conv_params=conv_params,
        fc_params=None,
        activation=activation,
        stride=stride,
        pooling=pooling,
        pool_every=pool_every
    )
    # #izvlacimo potrebnu informaciju
    conv_out_shape = conv_cache['conv_out_shape']
    flat_dim = int(np.prod(conv_out_shape))

    # ---- Fully connected params ----
    # dimenzije FC su sada [flat dim+ stare dimenzije]
    layer_dims = [flat_dim] + list(fc_layers)
    fc_params = {}  # recnik u kome cuvamo parametre
    for l in range(1, len(layer_dims)):  # iteriramo od 1 jer u sustini zelimo da dobijemo dimenzije svaki w ima dimenzije [prethodni sloj,sadasnji sloj]
        fan_in, fan_out = layer_dims[l-1], layer_dims[l]  # ovde zapisujemo te dimenzije
        # Xavier for tanh / softmax, He for ReLU
        if activation == 'tanh':  # odgovarajuce skaliranje
            w_scale = np.sqrt(1.0 / fan_in)
        else:
            w_scale = np.sqrt(2.0 / fan_in)
        fc_params[f'W{l}'] = np.random.randn(fan_out, fan_in) * w_scale  # pravimo parametre i zapisujemo ih u recnik
        fc_params[f'b{l}'] = np.zeros((fan_out, 1), dtype=float)

    return conv_params, fc_params # vracamo oba recnika
```

```python
def forward_propagation_conv_batch(X, conv_layers, conv_params, fc_params,
                                   activation='relu', stride=1, pooling=True, pool_every=2):
  """
    Šta se ovde dešava u suštini do sada kada bi provukli kroz forward propagaciju dobili bi samo jednu vrednost tacnije vrednost za samo jedan podatak
    Nas algoritam radi na samo jednom podatku sada je ideja da grupisemo sve podatke , naša predikcija finalno AL ne samo da treba da ima isti broj klasa kao
    i Y_train vec treba da ima i isti broj podataka

    Dakle do sada je bilo AL =(10,1) u multiklas (2,1) u binarz

    Nakon algoritma ce biti  AL=(10,m) u multiklas (2,m) u binary gde je m broj podataka u tom nasem trening skupu

    """




  m = X.shape[0]
  AL_cols, fc_caches, conv_caches = [], [], [] # cuvamo AL finalne Al vrednosti  kao i informacije o fc i conv  A, Z parametrima svakog sloja za svaki podatak

  use_pool = pooling and (pool_every is not None)

  for i in range(m):
      AL, fc_cache, conv_cache = forward_propagation_conv(
          x=X[i],
          conv_layers=conv_layers,
          conv_params=conv_params,
          fc_params=fc_params,
          activation=activation,
          stride=stride,
          pooling=use_pool,
          pool_every=pool_every
        )
        # obezbedi (C,1) po uzorku
      AL = np.asarray(AL)
      if AL.ndim == 1:
          AL = AL.reshape(-1, 1)
      elif AL.ndim == 2 and AL.shape[1] != 1 and AL.shape[0] == 1:
          AL = AL.T
      elif AL.ndim != 2 or AL.shape[1] != 1:
          raise ValueError(f"Per-sample AL expected (C,1), got {AL.shape}")

      AL_cols.append(AL)
      fc_caches.append(fc_cache)
      conv_caches.append(conv_cache)

  AL_batch = np.concatenate(AL_cols, axis=1)   # (C, m)
  return AL_batch, fc_caches, conv_caches
```

Backpropagacija klasicnog FCNN (Fully Connected Neural Networks)

```python
def backward_propagation(AL, Y, parametars, forward_cache, activation):
    grads = {} # recnik u kome skladistimo gradijente
    L = len(parametars) // 2 # broj slojeva ekvivalentan celobrojnom deljenju broju barametara jer imamo po dva u svakom sloju jedan w i jedan b
    m = AL.shape[1]  # broj instanci

    # Poslednji sloj:
    grads['dZ'+str(L)] = AL - Y
    grads['dW'+str(L)] = (1./m) * np.dot(grads['dZ'+str(L)], forward_cache['A'+str(L-1)].T)
    grads['db'+str(L)] = (1./m) * np.sum(grads['dZ'+str(L)], axis=1, keepdims=True)
    # svi ostali slojevi od L-1 do 1
    for l in reversed(range(1, L)):
        dA = np.dot(parametars['W'+str(l+1)].T, grads['dZ'+str(l+1)]) # dA po formuli
        if activation == 'tanh':
            dZ = dA * derivative_tanh(forward_cache['Z'+str(l)])# dz kao izvod u zavisnosti od aktivacione funkcije
        else:
            dZ = dA * derivative_relu(forward_cache['Z'+str(l)])
        grads['dZ'+str(l)] = dZ
        grads['dW'+str(l)] = (1./m) * np.dot(dZ, forward_cache['A'+str(l-1)].T)
        grads['db'+str(l)] = (1./m) * np.sum(dZ, axis=1, keepdims=True)
#dodavanje u recnik
    # dA0 potreban jer njega šaljemo Convolutivnim slojevima
    dA0 = np.dot(parametars['W1'].T, grads['dZ1'])   # dimenzije = (flat_dim, m)
    return grads, dA0
```

Cost funkcije odgovarajuce u zavisnosti da li je problem binarne ili multiklasne prirode

```python
def compute_cost(A, Y, eps=1e-12):
    A = np.asarray(A)
    Y = np.asarray(Y)

    if A.ndim == 1:
        A = A.reshape(1, -1)
    if Y.ndim == 1:
        Y = Y.reshape(1, -1)

    # numerička stabilnost
    A = np.clip(A, eps, 1.0 - eps)

    m = Y.shape[1]

    if A.shape[0] == 1 and Y.shape[0] == 1:
        # binarno: A i Y moraju biti (1, m)
        cost = -np.sum(Y * np.log(A) + (1 - Y) * np.log(1 - A)) / m
    else:
        # multiklasa: očekujemo (C, m) one-hot Y i softmax A
        assert A.shape == Y.shape, f"Multiclass očekuje A i Y istog oblika, dobio {A.shape} i {Y.shape}"
        cost = -np.sum(Y * np.log(A)) / m

    return float(cost)
```

```python
def bce_from_logits(Z, Y):
    """
    Binary cross-entropy from logits.
    Z: (1, m), Y: (1, m) with {0,1}
    """
    m = Y.shape[1]
    # stable softplus: log(1+exp(x))
    softplus = lambda x: np.log1p(np.exp(-np.abs(x))) + np.maximum(x, 0)
    L = Y * softplus(-Z) + (1 - Y) * softplus(Z)
    return float(np.sum(L) / m)

def cce_from_logits(Z, Y):
    """
    Multiclass cross-entropy from logits.
    Z: (C, m), Y: (C, m) one-hot
    """
    m = Y.shape[1]
    Z_shift = Z - np.max(Z, axis=0, keepdims=True)
    logsumexp = np.log(np.sum(np.exp(Z_shift), axis=0)) + np.max(Z, axis=0)
    L = -np.sum(Y * Z, axis=0) + logsumexp
    return float(np.sum(L) / m)
```

Backward Propagacija za jedan konvulucioni sloj

$
\frac{\partial L}{\partial Z^{[l]}}
= \frac{\partial L}{\partial A^{[l]}} \odot g'^{[l]}(Z^{[l]}),
$



```python
def backward_conv_one_layer(dA_curr,l,conv_cache,conv_params,activation='relu'):

 #  ulazne vrednosti trenutno dA, l-trenutni sloj, conv cache informacije o A[l-1] i dimenzijama filtera, conv_parametri-
  Z=conv_cache['Z'+str(l)]        # trenuntno  iz keša
  A_prev=conv_cache['A'+str(l-1)] # prethodno A iz keša
  meta=conv_cache['meta'+str(l)]  # dimenzije filtera nam trebaju
  f_h,f_w,s=meta['f_h'],meta['f_w'],meta['stride']

  H_in,W_in,C_in=A_prev.shape  # dimenzije prethodnog A
  H_out,W_out,K=Z.shape    # dimenzije trenutnog Z

  if activation == 'tanh':
    dZ=dA_curr*derivative_tanh(Z)  # dobijamo partial L/partial Z za l ti sloj
  else:
    dZ=dA_curr*derivative_relu(Z)

  dA_prev=np.zeros_like(A_prev) # dA sada racunamo po formuli

  dW_dict={} # Recnik u kome cuvamo gradijente hiperparametara

  for k in range(K):  # prolazimo kroz filtere
    wk=conv_params[l-1]['w'+str(k+1)] # uzimamo wk
    db_k=np.sum(dZ[:,:,k])# racunamo db
    dw_k=np.zeros_like(wk)# znamo da je dw iste dimenzije kao i wk

    oi=0  # zelimo da upisujemo odgovarajuce vrednosti na dw_k
    for i in range(0,H_in-f_h+1,s): # koliko se maksimalno pomeramo
      oj=0
      for j in range(0,W_in-f_w+1,s):
        patch=A_prev[i:i+f_h,j:j+f_w,:] # uzmemo isecak prethodnog A
        dw_k+=patch*dZ[oi,oj,k] # konvulucija

        #inverzna konvulucija
        dA_prev[i:i+f_h,j:j+f_w,:]+=wk*dZ[oi,oj,k] # transponovanje
        oj+=1
      oi+=1

    dW_dict['dw'+str(k+1)]=dw_k  # zapisujemo u recnik
    dW_dict['db'+str(k+1)]=db_k


  return dA_prev,dW_dict # vracamo prethodno dA i recnik
```

```python
def conv_backward(dA_from_fc,conv_layers,conv_params,conv_cache,activation='relu'):
  dA=dA_from_fc        # dobijamo transformisanu dakle prebacenu iz flatten oblika u onaj koji odgovara konvuluciji
  Lc=len(conv_layers)  # broj slojeva
  grads_conv={} # mesto gde belezimo gradijente

  for l in reversed(range(1,Lc+1)):  # iteriramo od prvog sloja do poslednjeg
    meta=conv_cache['meta'+str(l)] # vadimo informacije da znamo da li je vršeno pooling

    if meta['did_pool']: # ako jeste moramo vratiti informaciju nazad prisetimo se sa pooling smo ustvari smanjivali dimenziju matrice A ukoliko smo je
    #smanjili ocemo da je vratimo u originalnu dimenzionalnost
      dA=max_pooling_backward(dA,meta['pool']) # primena max_pooling_backward

    dA_prev,dW_dict=backward_conv_one_layer(dA,l,conv_cache,conv_params,activation) # primenjujemo prethodnu funkciju za sve slojeve

    grads_conv[l-1]=dW_dict # zamenjuje citav kod ispisan dole
    #grads_conv[l-1]={} # pamtimo gradijente sloja u recnik
    #K=conv_layers[l-1][0] #uzimamo broj filtera
    #for k in range(1,K+1): # opet je stvar notacije   l ide od 1 dakle za tkzv nulti sloj  dobijamo prve parametre zvane dw1
    #  grads_conv[l-1]['dw'+str(k)]=dW_dict['dw'+str(k)] # ubacujemo u odgovarajuci recnik vrednosti koje vaze ta taj sloj i sve filtere u tom sloju
    #  grads_conv[l-1]['db'+str(k)]=dW_dict['db'+str(k)]
    dA=dA_prev #nastavljamo iteraciju

  dA0=dA  # na kraju smos tigli do pocetnog ulaznog podatka koji bi trebao da je dimenzija naseg unosa podataka tj pocetne slike

  return grads_conv,dA0
```

```python
def backward_full(AL,Y,fc_params,fc_cache,conv_params,conv_cache,conv_layers,activation='relu'):
# u sustini dobijemo od forward propagacije  AL koji ima finalne predikcije i njega ubacimo u backward propagacija algoritam, koji nam kao rezultat vrati gradijente_fc i ravan dA0 koji trebamo
# reshapovati u odgovarajucu dimenziju AL poslednjeg konvolucionog sloja to postizemo tako sto pozovemo vrednost koju smo sacuvali u conv_cache
  grads_fc,dA0_flat=backward_propagation(AL,Y,fc_params,fc_cache,activation)
# nakon toga što prebacimo vektor ovu informaciju šaljemo nazad u konvoluciju backward koja nam vraca grads conv i dA0
  conv_out_shape=conv_cache['conv_out_shape']
  dA0_from_fc=dA0_flat.reshape(conv_out_shape)

  grads_conv,dA0=conv_backward(dA0_from_fc,conv_layers,conv_params,conv_cache,activation)

# ova funkcija radi za jedan podatak iz tog razloga je potrebna primena sledece funkcije
  return grads_fc,grads_conv,dA0
```

```python
def backward_full_batch(AL_batch, Y_batch,
                        fc_params, conv_params, conv_layers,
                        fc_caches, conv_caches,
                        activation='relu'):
    """
    Dakle slicna ideja kao i kod forward propagacije do sada sve sto smo opisali radi za pojedinacne slike i ubacivanje jednog podatka
    ideja je da sada napravimo funkciju koja ce raditi na poziv za sve podatke ulazni podaci su nam AL_batch - dimenzija (num_classes,num_examples)
    dakle nas broj primera je na lokaciji AL_batch[1] tj to nam govori koliko puta cemo pozivati algoritam.
    Što se tiče Y_batch on nam treba da bi izvukli rednost Y za svaki trenutni batch da bi izracunali u backward propagaciji poslednji sloj fully connected layera kao dZ=AL-Y
    conv layers za velicinu layera i vrednosti dimenzije filtera
    fc_caches
    """
    B = AL_batch.shape[1] # broj podataka tj primera

    # skladistimo vrednosti gradijenta fully connected layera
    grads_fc_acc = {} # # recnik
    Lf = len(fc_params)//2# broj slojeva
    for l in range(1, Lf+1):#
        grads_fc_acc['dW'+str(l)] = np.zeros_like(fc_params['W'+str(l)])# istih dimenzija su dw i w
        grads_fc_acc['db'+str(l)] = np.zeros_like(fc_params['b'+str(l)])#istih dimenzija su db i b

    grads_conv_acc = {} # konvulucioni
    for l in range(len(conv_layers)): # broj slojeva
        grads_conv_acc[l] = {} # za dati sloj pravimo recnik
        K = conv_layers[l][0] # broj filtera u datom sloju iz torke
        for k in range(1, K+1): # po filterima iteriramo
            grads_conv_acc[l]['dw'+str(k)] = np.zeros_like(conv_params[l]['w'+str(k)]) # u datom sloju numeracija krece od 1- L dodajemo parametre ali prisetimo se da prvi sloj odgoara idu parametri od 1 do k
            grads_conv_acc[l]['db'+str(k)] = 0.0 # parametrima w1 b1 i zabuna moze nastati sto smo u inicijalization funkciji ustvari numeraciju filtera krenuli  for k in 0- do K pa je filter 0 bio vezan za w1 b1
            # u suštini zapisi su ekvivalentni nema zabune za algoritam




    # iteriramo po primerima buduci da nas algoritam radi na jednim podacima odnosno jednoj slici pa prosledjujemo svaki ponaosob
    for b in range(B):
        AL_b = AL_batch[:,[b]] # uzimamo odgovarajucu AL vrednost
        fc_cache_b = fc_caches[b] # uzimamo odgovarajuce informacije o gradijentima
        conv_cache_b = conv_caches[b] # takodje za konvolucioni sloj
        # uzimamo isečak Y_b potreban za račun početnog dela petlje backpropagacije
        Y_b = Y_batch[:, [b]] if Y_batch.ndim == 2 else Y_batch[b]
# primenjujemo backpropagaciju na jednom podatku primenjuje se za sve i za fully connected i za conv slojeve
        grads_fc_b, grads_conv_b, _ = backward_full(  # uzmemo tj zapamtimo vrednosti gradijenata obe vrste NNa tj za fully conected slojeve i za conv slojeve
            AL_b, Y_b, fc_params, fc_cache_b,
            conv_params, conv_cache_b, conv_layers,
            activation=activation
        )
        # zatim prolazimo kroz kroz slojeve i zapisujemo sve vrednosti u recnik
        for l in range(1, Lf+1):
            grads_fc_acc['dW'+str(l)] += grads_fc_b['dW'+str(l)] # dakle dodajemo za ceo algoritam vrednost svakog pojedinacnog podatka tj dW od sakog pojedinacnog podatka tehnicki nasa grads fc_acc u jednom sloju
            grads_fc_acc['db'+str(l)] += grads_fc_b['db'+str(l)] # ce imati B puta ispisane vrednosti dW zbog toga radimo average da bi nakraju dobili smanjene dimenzije
        for l in range(len(conv_layers)):# prolazimo kroz konvulucione slojeve
          K = conv_layers[l][0]  # uzimamo broj filtera u datom sloju
          #gc_layer = grads_conv_b.get(l) or grads_conv_b.get(l+1)  # uzimamo informacije o gradijentima iz tog sloja ovde nam numeracija ide od 0 slojevima a kada smo belezili grads_conv_b
          gc_layer= grads_conv_b.get(l) # dakle preko backprop kada pratimo belezenje je takodje islo od nule tj prvi sloj ima oznaku 0 za parametre koji imaju oznaki dw1 db1.
          for k in range(1, K+1): # vadimo vrednosti gradijenata parametra u datom sloju za dati filter
            #key_w = 'w'+str(k) if 'w'+str(k) in gc_layer else 'dw'+str(k)
            #key_b = 'b'+str(k) if 'b'+str(k) in gc_layer else 'db'+str(k)
            grads_conv_acc[l]['dw'+str(k)] += gc_layer['dw'+str(k)]
            grads_conv_acc[l]['db'+str(k)] += gc_layer['db'+str(k)]

    # Kao što smo rekli za sada imamo u svakom sloju mestu ukupno B vrednosti dw i db za svaki fully connected i za svaki conv sloj pa radimo average ovih pravaca
    for l in range(1, Lf+1):  # od prvog do poslednjeg sloja radimo normalizaciju da tako kazem uzimamo srednju vrednost
        grads_fc_acc['dW'+str(l)] /= B
        grads_fc_acc['db'+str(l)] /= B
    for l in range(len(conv_layers)): # indeksiranje krece od nule do Lc-1 gde je 0 prvi sloj a poslednji Lc-1
        K = conv_layers[l][0] # za svaki sloj uzimamo odgovarajuci filter za cije parametre radimo average
        for k in range(1, K+1):
            grads_conv_acc[l]['dw'+str(k)] /= B
            grads_conv_acc[l]['db'+str(k)] /= B

    return grads_fc_acc, grads_conv_acc
```

```python
# one mini-batch gradient (averaged over the batch)
def compute_batch_grads_cnn(Xb, Yb, conv_layers, conv_params, fc_params,
                            activation='relu', stride=1, pooling=True, pool_every=2):
    AL_batch, fc_caches, conv_caches = forward_propagation_conv_batch(
        Xb, conv_layers, conv_params, fc_params,
        activation=activation, stride=stride, pooling=pooling, pool_every=pool_every
    )
    # This should already average over the batch inside
    grads_fc, grads_conv = backward_full_batch(
        AL_batch, Yb, fc_params, conv_params, conv_layers, fc_caches, conv_caches,
        activation=activation
    )
    # Return in the structure the SGD loop expects:
    return {'fc': grads_fc, 'conv': grads_conv}


# snapshot full gradient u (dataset-average per example, size-weighted across chunks)
def average_full_gradient_cnn(X, Y, conv_layers, snap_conv, snap_fc,
                              activation='relu', stride=1,
                              pooling=True, pool_every=2, full_bs=256):
    """
    Compute u = ∇R_n(w_k) at the snapshot params (snap_conv, snap_fc).
    X: (N,H,W,C), Y: (C,N)
    """
    n = X.shape[0]
    # make a flat zeros dict based on snapshot shapes
    u = zeros_like_flat(cnn_params_flat(snap_conv, snap_fc))
    seen = 0
    for s in range(0, n, full_bs):
        e = min(s + full_bs, n)
        Xb, Yb = X[s:e], Y[:, s:e]
        gb = compute_batch_grads_cnn(
            Xb, Yb, conv_layers, snap_conv, snap_fc,
            activation=activation, stride=stride,
            pooling=pooling, pool_every=pool_every
        )
        B = e - s
        for k in u:
            u[k] += gb[k] * B
        seen += B
    for k in u:
        u[k] /= max(1, seen)
    return u
```

```python


def train_val_split_flexible(X, Y, val_ratio=0.2, seed=None):
    """
    Želimo da podelimo x i y u train val odnos i za fully connnected layers i za CNN layers
    dakle ubacujemo X_train, Y_train i kao rezulat dobijamo manji X_train, Y_train i X_val Y_val
   X za fnn je oblika (features,N) a za X je oblika (N,H,W,C) ili (N,H,W)

    """
    if seed is not None:
        np.random.seed(seed)# koristi se za slucajan izbor podataka koji ce pripadovati validacionom skupu

    # detektujemo broj podataka naime prisetimo se kako smo opisali flat vektor A kada smo ga prebacivali
#        dA0 = np.dot(parametars['W1'].T, grads['dZ1'])   # dimenzije = (flat_dim, m) backpropagacijom fully connected sloja
 #   return grads, dA0
#dimenzije su bile flat_dim tj ovo su osobine kao pixeli, a m je broj podataka dakle nase N je na drugoj poziciji odnosno ako je x dvodimenzijalno a
# to se desava samo u FC slojevima onda se informacija o broju podataka nalazi na drugoj poziciji tj indeks [1]
    if X.ndim == 2:
        # vadimo taj indeks
        N = X.shape[1]
        sample_axis_X = 1
    # u suprotnom ako imamo 3/4 dimenzije pocetnog vektora onda je na prvoj poziciji
    elif X.ndim in (3, 4):
        # CNN: (N, H, W[, C])
        N = X.shape[0]
        sample_axis_X = 0
    else:
        raise ValueError(f"Unsupported X.ndim={X.ndim}")

#u sustnini za nas problem je bio potreban samo ovaj drugi deo posto mi odma ubacujemo sliku a ne vektor
#  ovo napravljeno da radi uopsteno i za slucaj da radimo obicne neuronske mreze.

    # proveravamo dimenzionalnost y u suštini mi radimo sa y (oblika br_klasa,br_primera) dakle dve dimenzije i na drugoj je broj brj primera
    if Y.ndim == 1: #
        # ako je jedna dimenzija oblika je N ovo je broj primera
        assert Y.shape[0] == N
        index_Y_on_axis1 = False  #
    elif Y.ndim == 2: # ako su dve dimenzije
        #  i ako je na drugoj n onda uzimamo sve piksele tj informacije te slike za dati index
        if Y.shape[1] == N:
            index_Y_on_axis1 = True  # oznacava dakle da je indeksoanje na jedinici  koristimo [:,inx]
        elif Y.shape[0] == N: # ako je na nultom indeksu šejpa tj ako ide (primeri,klase)
            index_Y_on_axis1 = False  # koristimo [idx, :]
        else:
            raise ValueError("Y must have N on axis 0 or axis 1.")
    else:
        raise ValueError(f"Unsupported Y.ndim={Y.ndim}")

    # Indices
    idx = np.random.permutation(N) #slučajni preraspodela brojeva od 1 do N permutacija bez ponavljanja
    v = int(N * val_ratio)  # broj koliko je podataka u val skupu
    val_idx, train_idx = idx[:v], idx[v:]  # uzimamo prvih v slucajno rapsoredjenih podataka za indekse validacijonog skupa a ostalo za indekse trening skupa

    # u zavisnosti od toga da li je dimenzija vektora ili tenzora
    if sample_axis_X == 1:
        X_train = X[:, train_idx]
        X_val   = X[:, val_idx]
    else:
        X_train = X[train_idx]
        X_val   = X[val_idx]

    # Takodje razdvajamo slucaj ovaj prvi kod nas nije zadovoljen jer je y oblika num_classes,num_examples
    if Y.ndim == 1:
        Y_train = Y[train_idx]
        Y_val   = Y[val_idx]
    else: # pa je pitanje da li je na drugom indeksu num_examples
        if index_Y_on_axis1:# jeste
            Y_train = Y[:, train_idx] # pravilan ispis (num_classes,suma ndeks)
            Y_val   = Y[:, val_idx]
        else: # ovo je netacan zapis kod nas ne ispisuje se
            Y_train = Y[train_idx, :]
            Y_val   = Y[val_idx, :]

    return X_train, Y_train, X_val, Y_val  # vracamo x_train , y train xval yval
```

```python
def model_sgd_cnn(
    X_train, Y_train,                  # X: (N,H,W,C), Y: (C,N) one-hot / binary
    conv_layers, fc_layers,            # architecture
    conv_params=None, fc_params=None,  # if None -> init
    X_val=None, Y_val=None,
    activation='relu', stride=1, pooling=True, pool_every=2,
    lr=0.01, batch_size=64, epochs=5,
    weight_decay=1e-4, momentum=0.9,   # set momentum=0.0 to disable
    seed=1, report_every=1
):
    """
    prvi parametri koje ubacujemo su jasni i vec puno puta spominjani, u sustini mi
    ne koristimo pravi sgd osim ukoliko nije naglasen batch_size=1 u suprotnom koristimo mini
    batch gradient descent sto je dobar kompromis brzine i memorije
    jer je full batch brzo konvergentan spor komputacijski zahtevan za racun i memorijski ne toliko
    dok je sgd sporiji komputacijski jeftin ali memorijski jako skup
    broj epoha je ocigledno znacenje

    weight decay L2 regaularizacija koja penalizuje velika w i pomaze u generalizaciji,
    Momentum filtrira šum i ubrzava kretanje duž stabilnih pravaca
    report every na koliko epoha da izvestavamo
    """

    if seed is not None:# ukoliko nije postavljeno uzimamo slucajno
      np.random.seed(seed)


    N, H, W, C = X_train.shape # pocetne dimenzije dakle (N,H,W,C) u nasem slucaju H,W,C 64,64,3 gde je N broj podataka
    assert Y_train.shape[1] == N #osiguravamo da je # Y dimenzija (klase,broj podataka)

    # ----inicijalizacija parametara ako je potrebna
    if conv_params is None or fc_params is None:
        conv_params, fc_params = initialize_parameters(
            conv_layers=conv_layers,
            input_shape=(H, W, C),
            fc_layers=fc_layers,
            pooling=(pool_every is not None) and pooling,
            pool_every=pool_every,
            activation=activation, stride=stride, seed=seed
        )


    def batched_metrics(X, Y, bs=256):# skup dimenzija slika
        M = X.shape[0] # ustvari N tj broj primera
        correct, total_loss, seen = 0, 0.0, 0  # postavljamo brj tacno pogodjenih, sumu gubitka preko svih i koliko smo primera obradili na nulu
        for s in range(0, M, bs): # iteriramo po mini-batchevima velicine bs
            e = min(s+bs, M) # s:e su indeksi mini batcha e je gornja granica ne prelazimo M
            ALb_list, _, _ = forward_propagation_conv_batch(# napred prosledjujemo baš ovaj bač
                X[s:e], conv_layers, conv_params, fc_params,
                activation=activation, stride=stride,
                pooling=(pool_every is not None) and pooling, pool_every=pool_every
            )# vracamo AL_vrednosti
            ALb = np.concatenate([np.asarray(AL).reshape(-1, 1) for AL in ALb_list],# skvizuje o
                                  axis=1)  # (C, B)
            Yb  = Y[:, s:e] # uzimamo isti isecak iz trening skupa

            # accuracy
            if (ALb.ndim == 1) or (ALb.shape[0] == 1):  # ako je binarna
                preds = (np.ravel(ALb) >= 0.5).astype(int)# vid predikcije
                true  = np.ravel(Yb).astype(int) if Yb.ndim == 1 else np.argmax(Yb, axis=0)
            else:                                       # ako je multiklas
                preds = np.argmax(ALb, axis=0)
                true  = np.argmax(Yb,  axis=0)
            correct += np.sum(preds == true) # ako smo pogodili povecavamo broj tacnih

            # loss
            total_loss += compute_cost(ALb, Yb) * (e - s) # funkcija gubitka
            seen += (e - s) # pregledanih

        return correct / max(1, M), total_loss / max(1, seen) #tr_acc tr_loss

    # pravimo nula matricu ubrzivace za momentum koje su istih dimenzija kao i sami parametri buduci da se
    def zeros_like_params(conv_p, fc_p):# na njih dodaju
        v_conv = {l: {k: (np.zeros_like(v) if isinstance(v, np.ndarray) else 0.0)
                      for k, v in conv_p[l].items()} for l in conv_p}
        v_fc   = {k: np.zeros_like(v) for k, v in fc_p.items()}
        return v_conv, v_fc
# ona je nula jer u prvoj iteraciji ubrzavanja nema
    v_conv, v_fc = zeros_like_params(conv_params, fc_params)

    # ---- prolazimo kroz epohe i treniramo
    for epoch in range(1, epochs+1):
        # shuffle indices
        perm = np.random.permutation(N)# permutacija slucajan izbor redosleda
        for s in range(0, N, batch_size): # od 0 do N sa korakom batch_size
            e = min(s+batch_size, N)# gornja granica bathca
            idx = perm[s:e] # podksup nasumicnih indeksa
            Xb, Yb = X_train[idx], Y_train[:, idx]

            # racunamo gradijente za ovaj bath kao resenje dobijamo recnik
            grads = compute_batch_grads_cnn(
                Xb, Yb, conv_layers, conv_params, fc_params,
                activation=activation, stride=stride,
                pooling=(pool_every is not None) and pooling, pool_every=pool_every
            )
            # ovaj recnik sadrzi u sebi dva recnika jedan pod nazivom fc drugi pod nazivom conv.
            # saki od recnika sadrzi svoje gradijente

            grads_conv = grads['conv'] if 'conv' in grads else grads[0]
            grads_fc   = grads['fc']   if 'fc'   in grads else grads[1]

            # L2 weight decay: dakle u slucaju regularizacije
            if weight_decay != 0.0: # ako imamo regularizacioni hiper parametar
                for l in conv_params:# idemo po kljucevima nama je conv_params oblika {0:{....},1:{....}..}
                    for k, w in conv_params[l].items():# ovde pristupimo odgovarajucim k i w vrednostima
                        if isinstance(w, np.ndarray) and w.ndim >= 1 and k.startswith('w'):
                          gkey='dw'+k[1:] # za odgovarajuci kernel
                          assert grads_conv[l][gkey].shape==w.shape
                          grads_conv[l][gkey] = grads_conv[l][gkey] + weight_decay * w #updajtujemo parametre sa regularizacijom
                Lf = len(fc_layers)
                for l in range(1, Lf+1):# po slojevima FC
                    grads_fc[f'dW{l}'] = grads_fc[f'dW{l}'] + weight_decay * fc_params[f'W{l}'] # updajte parametara sa regularizacijom

            # momentum + SGD tacnije batch
            mu = momentum # ubrzanje koje imamo
            # momentum za conv
            for l in conv_params: # prolayimo opet kao gore
              for name, param in conv_params[l].items():   # opet pristupamo kao gore
                if name.startswith('w'):
                  gkey = 'dw' + name[1:]               # preimenujemo gkey u dW1 iz w1 itd
                elif name.startswith('b'):
                  gkey = 'db' + name[1:]               # takodje ovde samo za db
                else:
                  continue

                g = grads_conv[l][gkey]

                # primena momentuma
                v_conv[l][name] = mu * v_conv[l][name] + g

                # update parametra
                conv_params[l][name] = conv_params[l][name] - lr * v_conv[l][name]

            # isto za fc
            Lf = len(fc_layers)
            for l in range(1, Lf+1):
                for nameW, nameb, dW, db in [
                    (f'W{l}', f'b{l}', f'dW{l}', f'db{l}')
                ]:
                    v_fc[nameW] = mu * v_fc[nameW] + grads_fc[dW]
                    v_fc[nameb] = mu * v_fc[nameb] + grads_fc[db]
                    fc_params[nameW] = fc_params[nameW] - lr * v_fc[nameW]
                    fc_params[nameb] = fc_params[nameb] - lr * v_fc[nameb]

        # kad se zavrsi epoha  ukoliko je deljiva sa report_every pravimo izvestaj
        if (epoch % report_every) == 0:
            tr_acc, tr_loss = batched_metrics(X_train, Y_train, bs=batch_size)
            if (X_val is not None) and (Y_val is not None):
                va_acc, va_loss = batched_metrics(X_val, Y_val, bs=batch_size)
                print(f"[epoch {epoch}] "
                      f"train_acc: {tr_acc:.4f}  train_loss: {tr_loss:.4f}  "
                      f"val_acc: {va_acc:.4f}    val_loss: {va_loss:.4f}")
            else:
                print(f"[epoch {epoch}] train_acc: {tr_acc:.4f}  train_loss: {tr_loss:.4f}")

    return conv_params, fc_params
```

```python
def model_svrg_cnn(
    X_train, Y_train,                  # X: (N,H,W,C), Y: (C,N)
    conv_layers, fc_layers,            # architecture
    conv_params=None, fc_params=None,  # if None -> init
    X_val=None, Y_val=None,
    activation='relu', stride=1, pooling=True, pool_every=2,
    lr=0.01, batch_size=128, full_bs=512,
    outer_loops=5, m=None,             # m = #inner steps per outer loop; default N//batch
    weight_decay=1e-4, avg_inner=True, seed=1, report_every=1
):
    """
    SVRG for your manual CNN, fully compatible with your working SGD code:
    - Uses the SAME param layout and the SAME grad dicts as your SGD loop.
    - No flat/stacked renaming; keys in grads must match keys in params.
    - L2 (weight decay) applied to W-keys like in your SGD (k.startswith('w')) and FC dW.
    """

    import numpy as np
    np.random.seed(seed)

    N, H, W, C = X_train.shape
    assert Y_train.shape[1] == N, "Y_train must be (num_classes, N)."

    # ---------- init params ----------
    if conv_params is None or fc_params is None:
        conv_params, fc_params = initialize_parameters(
            conv_layers=conv_layers,
            input_shape=(H, W, C),
            fc_layers=fc_layers,
            pooling=(pool_every is not None) and pooling,
            pool_every=pool_every,
            activation=activation, stride=stride, seed=42
        )

    # ---------- small helpers on NESTED dicts ----------
    def clone_params(cp, fp):
        cc = {l: {k: (v.copy() if isinstance(v, np.ndarray) else v) for k, v in cp[l].items()} for l in cp}
        ff = {k: (v.copy() if isinstance(v, np.ndarray) else v) for k, v in fp.items()}
        return cc, ff

    def zeros_like_grads(g_conv, g_fc):
        zc = {l: {k: np.zeros_like(v) for k, v in g_conv[l].items()} for l in g_conv}
        zf = {k: np.zeros_like(v) for k, v in g_fc.items()}
        return zc, zf

    def add_nested(a_conv, a_fc, b_conv, b_fc, alpha=1.0):
        out_c = {l: {} for l in a_conv}
        for l in a_conv:
            for k in a_conv[l]:
                out_c[l][k] = a_conv[l][k] + alpha * b_conv[l][k]
        out_f = {}
        for k in a_fc:
            out_f[k] = a_fc[k] + alpha * b_fc[k]
        return out_c, out_f

    def sub_nested(a_conv, a_fc, b_conv, b_fc):
        return add_nested(a_conv, a_fc, b_conv, b_fc, alpha=-1.0)

    def scale_nested(a_conv, a_fc, alpha):
        out_c = {l: {k: alpha * v for k, v in a_conv[l].items()} for l in a_conv}
        out_f = {k: alpha * v for k, v in a_fc.items()}
        return out_c, out_f

    def apply_weight_decay_inplace(grads_conv, grads_fc, conv_p, fc_p, lam):
        if lam == 0.0:
            return
        # conv: only weights (keys starting with 'w'), like your SGD
        for l in conv_p:
            for k, w in conv_p[l].items():
                if isinstance(w, np.ndarray) and w.ndim >= 1 and k.lower().startswith('w'):
                    grads_conv[l][k] = grads_conv[l][k] + lam * w
        # fc: dWl += lam * Wl (your SGD naming)
        Lf = len(fc_layers)
        for l in range(1, Lf + 1):
            grads_fc[f'dW{l}'] = grads_fc[f'dW{l}'] + lam * fc_params[f'W{l}']

    def compute_batch_grads(Xb, Yb, conv_p, fc_p):
        g = compute_batch_grads_cnn(
            Xb, Yb, conv_layers, conv_p, fc_p,
            activation=activation, stride=stride,
            pooling=(pool_every is not None) and pooling, pool_every=pool_every
        )
        # normalize to dicts grads_conv, grads_fc like in your SGD
        if isinstance(g, dict):
            grads_conv = g['conv'] if 'conv' in g else g[0]
            grads_fc   = g['fc']   if 'fc'   in g else g[1]
        else:
            grads_conv, grads_fc = g
        return grads_conv, grads_fc

    # simple metrics (same as in your SGD)
    def compute_cost_from_probs(A, Y, eps=1e-12):
        A = np.clip(A, eps, 1.0 - eps)
        m = Y.shape[1]
        if A.shape[0] == 1:
            return float(-(1.0/m) * np.sum(Y*np.log(A) + (1 - Y)*np.log(1 - A)))
        else:
            return float(-(1.0/m) * np.sum(Y * np.log(A)))

    def batched_metrics(X, Y, bs=256):
        M = X.shape[0]
        correct, total_loss, seen = 0, 0.0, 0
        for s in range(0, M, bs):
            e = min(s+bs, M)
            ALb, _, _ = forward_propagation_conv_batch(
                X[s:e], conv_layers, conv_params, fc_params,
                activation=activation, stride=stride,
                pooling=(pool_every is not None) and pooling, pool_every=pool_every
            )
            ALb = np.stack(ALb, axis=1).squeeze(-1)
            Yb  = Y[:, s:e]

            if (ALb.ndim == 1) or (ALb.shape[0] == 1):
                preds = (np.ravel(ALb) >= 0.5).astype(int)
                true  = np.ravel(Yb).astype(int) if Yb.ndim == 1 else np.argmax(Yb, axis=0)
            else:
                preds = np.argmax(ALb, axis=0)
                true  = np.argmax(Yb,  axis=0)
            correct += np.sum(preds == true)

            total_loss += compute_cost_from_probs(ALb, Yb) * (e - s)
            seen += (e - s)

        return correct / max(1, M), total_loss / max(1, seen)

    # ---------- SVRG settings ----------
    n = N
    if m is None:
        m = max(1, n // batch_size)  # inner steps per outer loop

    best_val = -np.inf
    best_conv, best_fc = None, None

    # ---------- outer SVRG cycles ----------
    for k in range(1, outer_loops + 1):
        # snapshot params
        snap_conv, snap_fc = clone_params(conv_params, fc_params)

        # full gradient at snapshot: u = (1/N) * sum_b grad_b(snap)
        # computed in chunks of size full_bs
        u_conv, u_fc = None, None
        seen = 0
        for s in range(0, N, full_bs):
            e = min(s + full_bs, N)
            conv_gb, fc_gb = compute_batch_grads(X_train[s:e], Y_train[:, s:e], snap_conv, snap_fc)

            # average over the full batch (plain sum; we normalize at the end by N)
            if u_conv is None:
                # init accumulators with zeros like first grad
                u_conv, u_fc = zeros_like_grads(conv_gb, fc_gb)
            # accumulate
            u_conv, u_fc = add_nested(u_conv, u_fc, conv_gb, fc_gb, alpha=(e - s))
            seen += (e - s)

        # finalize average
        u_conv, u_fc = scale_nested(u_conv, u_fc, 1.0 / float(seen))

        # ---------- inner loop ----------
        # (Option B): average inner iterates
        import numpy as np

        def zeros_like_params(conv_p, fc_p):
          acc_c = {l: {k: (np.zeros_like(v) if isinstance(v, np.ndarray) else 0.0)
                 for k, v in conv_p[l].items()}
             for l in conv_p}
          acc_f = {k: (np.zeros_like(v) if isinstance(v, np.ndarray) else 0.0)
             for k, v in fc_p.items()}
          return acc_c, acc_f

        if avg_inner:
          acc_conv, acc_fc = zeros_like_params(conv_params, fc_params)


        for j in range(m):
            idx = np.random.randint(0, n, size=(batch_size,))
            Xb, Yb = X_train[idx], Y_train[:, idx]

            # current grads
            g_curr_conv, g_curr_fc = compute_batch_grads(Xb, Yb, conv_params, fc_params)
            # snapshot grads (same minibatch, but at snap params)
            g_snap_conv, g_snap_fc = compute_batch_grads(Xb, Yb, snap_conv, snap_fc)

            # variance reduction: g̃ = g_curr - g_snap + u
            tmp_conv, tmp_fc = sub_nested(g_curr_conv, g_curr_fc, g_snap_conv, g_snap_fc)
            g_vr_conv, g_vr_fc = add_nested(tmp_conv, tmp_fc, u_conv, u_fc, alpha=1.0)

            # add weight decay (same convention as your SGD)
            apply_weight_decay_inplace(g_vr_conv, g_vr_fc, conv_params, fc_params, weight_decay)

            # SGD step with g̃ (no momentum in vanilla SVRG)
            for l in conv_params:
                for k2 in conv_params[l]:
                    conv_params[l][k2] = conv_params[l][k2] - lr * g_vr_conv[l][k2]
            Lf = len(fc_layers)
            for l in range(1, Lf + 1):
                for nameW, nameb, dW, db in [(f'W{l}', f'b{l}', f'dW{l}', f'db{l}')]:
                    fc_params[nameW] = fc_params[nameW] - lr * g_vr_fc[dW]
                    fc_params[nameb] = fc_params[nameb] - lr * g_vr_fc[db]

            # accumulate for option-B averaging
            if avg_inner:
                for l in conv_params:
                    for k2 in conv_params[l]:
                        acc_conv[l][k2] += conv_params[l][k2]
                for k2 in fc_params:
                    acc_fc[k2] += fc_params[k2]

        # end inner: assign averaged params if requested
        if avg_inner:
            for l in conv_params:
                for k2 in conv_params[l]:
                    conv_params[l][k2] = acc_conv[l][k2] / float(m)
            for k2 in fc_params:
                fc_params[k2] = acc_fc[k2] / float(m)

        # ---------- metrics & best snapshot ----------
        tr_acc, tr_loss = batched_metrics(X_train, Y_train, bs=256)
        if (X_val is not None) and (Y_val is not None):
            va_acc, va_loss = batched_metrics(X_val, Y_val, bs=256)
            if va_acc > best_val:
                best_val = va_acc
                best_conv, best_fc = clone_params(conv_params, fc_params)
            if (k % report_every) == 0:
                print(f"[SVRG outer {k}] "
                      f"train_acc: {tr_acc:.4f}  train_loss: {tr_loss:.4f}  "
                      f"val_acc: {va_acc:.4f}    val_loss: {va_loss:.4f}")
        else:
            if (k % report_every) == 0:
                print(f"[SVRG outer {k}] train_acc: {tr_acc:.4f}  train_loss: {tr_loss:.4f}")

    return (best_conv, best_fc) if (X_val is not None and Y_val is not None) else (conv_params, fc_params)
```
